```python
# see -- https://towardsdatascience.com/machine-learning-nlp-text-classification-using-scikit-learn-python-and-nltk-c52b92a7c73a
from sklearn.datasets import fetch_20newsgroups
twenty_train = fetch_20newsgroups(subset='train', shuffle=True)
str_list = twenty_train.data
target_classification_for_str_list = twenty_train.target
target_classification_names = twenty_train.target_names
print('We have %d data'%(len(str_list)))
```

```python
print(type(str_list))
```

```python
# Extracting features from str_list
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(np.array(str_list))
```

```python
print(X_train_counts.shape)
print('sklearn extracted %d word features from %d string.'%(X_train_counts.shape[0], X_train_counts.shape[1]))
print('each row is a feature')
print('each column is a data')
print('feature of data[0]: %s (n=%d)' % (np.array_str(X_train_counts.getcol(0).toarray().flatten() ), len(X_train_counts.getcol(0).toarray().flatten()) ) )
```

```python
from sklearn import cluster, datasets
iris = datasets.load_iris()
X_iris = iris.data
y_iris = iris.target
```

```python
print(X_iris.shape) # ncol=4, nrow=150, number of dimension is considered 4
```

```python
k_means = cluster.KMeans(n_clusters=4)
k_means.fit(X_iris)
print(k_means.labels_[::10])
print(y_iris[::10])
```

```python
import matplotlib.pyplot as plt
print(len(X_iris[:,0]))
fig = plt.figure()
ax1 = fig.add_subplot(1,3,1)
ax1.scatter(X_iris[:,0], X_iris[:,1], c=k_means.labels_)
ax2 = fig.add_subplot(1,3,2)
ax2.scatter(X_iris[:,0], X_iris[:,2], c=k_means.labels_)
ax3 = fig.add_subplot(1,3,3)
ax3.scatter(X_iris[:,0], X_iris[:,3], c=k_means.labels_)
```
